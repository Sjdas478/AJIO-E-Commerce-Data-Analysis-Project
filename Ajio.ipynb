# Customer Purchase Behavior

Customer Age Distribution

```python
plt.figure(figsize=(8,5))
sns.histplot(customer['Age'], bins=10, kde=True)
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.tight_layout()
plt.show()
```

Gender-wise Purchase Count

```python
orders_customer=orders.merge(customer,on='C_ID')
gender_order_counts = orders_customer['Gender'].value_counts()
print(gender_order_counts)

plt.figure(figsize=(4,4))
plt.pie(gender_order_counts, labels=gender_order_counts.index, autopct='%1.1f%%', startangle=90, colors=['#ADD8E6', '#FFB6C1'])
plt.title('Gender-wise Purchase Count (by Number of Orders)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

 Top 10 Cities by Number of Orders

```python
city_wise = orders_customer['City'].value_counts().head(10)
print(city_wise)

plt.figure(figsize=(10,5))
sns.barplot(x=city_wise.values, y=city_wise.index)
plt.title('Top 10 Cities by Number of Orders')
plt.xlabel('Number of Orders')
plt.ylabel('City')
plt.tight_layout()
plt.show()
```

 Top 5 States by Number of Orders

```python
State_wise = orders_customer['State'].value_counts().head()
print(State_wise)

plt.figure(figsize=(10,5))
sns.barplot(x=State_wise.values, y=State_wise.index)
plt.title('Top 10 States by Number of Orders')
plt.xlabel('Number of Orders')
plt.ylabel('States')
plt.tight_layout()
plt.show()
```

Top 10 Repeat Customers (by number of orders)

```python
repeat_customers = orders['C_ID'].value_counts().head(10)

plt.figure(figsize=(8,5))
sns.barplot(x=repeat_customers.values, y=repeat_customers.index,hue=plt.ylabel,palette='magma',legend=False)
plt.title('Top 10 Repeat Customers')
plt.xlabel('Number of Orders')
plt.ylabel('Customer ID')
plt.tight_layout()
plt.show()
```

Active vs Inactive Customers

```python
active_customer_ids = orders['C_ID'].unique()
inactive_customers = customer[~customer['C_ID'].isin(active_customer_ids)]

print("Total customers:", len(customer))
print("Active customers:", len(active_customer_ids))
print("Inactive customers:", len(inactive_customers))
```

# Product & Category Performance

```python
orders_product=orders.merge(products,on='P_ID')
```

```python
orders_product['Revenue'] = orders_product['Price'] * orders_product['Qty'] * (1 - orders_product['Discount']/100)
```

Top-Selling Products by Quantity

```python
top_product = orders_product.groupby('P_Name')['Qty'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_product.values, y=top_product.index, hue=plt.ylabel,palette='crest',legend=False)
plt.title('Top 10 Best-Selling Products by Quantity')
plt.xlabel('Total Quantity Sold')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

Top 5-Selling Categories by Quantity

```python
category_sales = orders_product.groupby('Category')['Qty'].sum().sort_values(ascending=False).head()

plt.figure(figsize=(5,3))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title('Top 5 Categories by Quantity Sold')
plt.xlabel('Quantity')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

Category-wise Revenue Share

```python
category_revenue = orders_product.groupby('Category')['Revenue'].sum()

plt.figure(figsize=(4,4))
plt.pie(category_revenue, labels=category_revenue.index, autopct='%1.1f%%', startangle=140)
plt.title('Category-wise Revenue Share')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

Top 5 Companies by Quantity

```python
revenue_by_company = orders_product.groupby('Company_Name')['Revenue'].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(8,4))
sns.barplot(x=revenue_by_company.index, y=revenue_by_company.values)
plt.title('Top 5 Companies by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Company')
plt.tight_layout()
plt.show()
```

Gender-wise Product Sales (Based on Quantity)

```python
gender_sales = orders_product.groupby('Gender')['Qty'].sum()
plt.figure(figsize=(4,4))
plt.pie(gender_sales, labels=gender_sales.index, autopct='%1.1f%%', startangle=90, colors=['#ADD8E6', '#FFB6C1', '#D3D3D3'])
plt.title('Gender-wise Product Sales (by Quantity)')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

# Return Pattern Analysis

```python
returns_orders = returns.merge(orders, on='Or_ID')
returns_full = returns_orders.merge(products, on='P_ID')
```

Top Return Reasons

```python
return_reasons = returns['Reason'].value_counts()
print(return_reasons)
plt.figure(figsize=(4,4))
plt.pie(return_reasons, labels=return_reasons.index, autopct='%1.1f%%', startangle=90)
plt.title('Return Reasons Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()
```

 Return Count by Category

```python
returns_by_category = returns_full['Category'].value_counts()

plt.figure(figsize=(8,5))
sns.barplot(x=returns_by_category.values, y=returns_by_category.index)

plt.title('Returns by Product Category')
plt.xlabel('Number of Returns')
plt.ylabel('Category')
plt.tight_layout()
plt.show()
```

Return Approval vs Rejection

```python
status_counts = returns['Return_Refund'].value_counts()
print(status_counts)

plt.figure(figsize=(4,3))
sns.barplot(x=status_counts.index, y=status_counts.values, hue=status_counts.index,palette='Set2')
plt.title('Return Refund Status')
plt.xlabel('Status')
plt.ylabel('Count')
plt.tight_layout()
plt.show()

```

Delivery Partner Involvement in Returns

```python
returns_with_dp = returns.merge(orders[['Or_ID', 'DP_ID']], on='Or_ID')
returns_with_dp = returns_with_dp.merge(delivery, on='DP_ID')

dp_returns = returns_with_dp['DP_name'].value_counts()

plt.figure(figsize=(6,5))
sns.barplot(x=dp_returns.values, y=dp_returns.index, hue=dp_returns.index,palette='Purples')
plt.title('Returns by Delivery Partner')
plt.xlabel('Number of Returns')
plt.ylabel('Delivery Partner')
plt.tight_layout()
plt.show()

```

# Delivery Partner Performance

```python
orders_delivery = orders.merge(delivery, on='DP_ID')
ratings_merged = ratings.merge(orders[['Or_ID', 'DP_ID']], on='Or_ID')
ratings_full = ratings_merged.merge(delivery, on='DP_ID')
```

Average Delivery Rating per Partner

```python
avg_delivery_rating = ratings_full.groupby('DP_name')['Delivery_Service_Rating'].mean().sort_values(ascending=False)

plt.figure(figsize=(6,4))
sns.barplot(x=avg_delivery_rating.index, y=avg_delivery_rating.values,hue=avg_delivery_rating.index, palette='Blues_r')
plt.title('Average Delivery Rating per Partner')
plt.xlabel('Average Rating')
plt.ylabel('Delivery Partner')
plt.tight_layout()
plt.show()

```



Partner Ratings vs Percent Cut

```python

plt.figure(figsize=(4,4))
sns.scatterplot(data=delivery, x='Percent_Cut', y='DP_Ratings', s=100, color='teal')
plt.title('Delivery Partner Rating vs Commission Cut')
plt.xlabel('Commission % Cut')
plt.ylabel('Average Delivery Rating')
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Transaction & Payment Analysis

```python
orders_products = orders.merge(products[['P_ID', 'Price']], on='P_ID', how='left')

```

```python
orders_products['Or_Price'] = orders_products['Qty'] * orders_products['Price']
```

```python
orders_products['Discount'] = pd.to_numeric(orders_products['Discount'], errors='coerce')
```

```python
orders_products['Coupon_Status'] = orders_products['Coupon'].apply(
    lambda x: 'No Coupon' if pd.isna(x) or x == 'No Coupon' else 'Used Coupon'
)
```

Total Revenue

```python
Total_Revenue=orders_product["Revenue"].sum()
print(f'Total_Revenue is {Total_Revenue}')
```

Average Order Value (AOV)

```python
order_value = orders_product.groupby('Or_ID')['Revenue'].sum()


aov = order_value.mean()
print(f"Average Order Value: {aov:,.2f}")
```

Most preferred Payment Method

```python
payment_counts = transactions['Transaction_Mode'].value_counts()

plt.figure(figsize=(4,4))
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Transaction Mode Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()


```

Coupon Usage Distribution

```python
coupon_usage = orders_products['Coupon_Status'].value_counts()

plt.figure(figsize=(4,4))
plt.pie(
    coupon_usage.values,
    labels=coupon_usage.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=['#66b3ff', '#ff9999']
)
plt.title('Coupon Usage Distribution')
plt.axis('equal')
plt.tight_layout()
plt.show()

```

Total Order Value by Coupon Status

```python
coupon_order_value = orders_products.groupby('Coupon_Status')['Or_Price'].sum()
print(coupon_order_value)
```

How Many People used Coupons

```python
coupon_users = orders_products[orders_products['Coupon'].notna() & (orders_products['Coupon'] != 'No Coupon')]
```

```python
coupon_users = coupon_users['C_ID'].count()
print("Number of  customers who used a coupon:", coupon_users)
```
